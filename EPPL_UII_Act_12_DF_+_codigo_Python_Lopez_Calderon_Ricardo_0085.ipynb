```python
# calculando la diferecia de valores usando if else
# datos de entrada A,B
# mostrar B si es falso
# mostrar A si es verdadero

A = float(input("Ingrese el primer valor (A): "))
B = float(input("Ingrese el segundo valor (B): "))

if A > B:
  M = A
else:
  M = B

print("El mayor es:", M)
print("Fin del programa")
print("Programa realizado por Lopez Ricardo NL = 29")
```
